```python
from collections import Counter

# Use Counter to get the 10 most common words
top_20_words = Counter(list).most_common(20)

# Print the top 10 words
print("Top 10 most occurring words:")
for word, count in top_20_words:
    print(f"{word}: {count}")
```

```python
uni = set (list)
print (uni)
```

```python
type (uni)
```

```python
d = {}
for i in uni:
  d[i] = list.count (i)
```

```python
print (d)
```

```python
# Sort the dictionary items by count in descending order and get the top 10
top_10_words_from_d = sorted(d.items(), key=lambda item: item[1], reverse=True)[:10]

# Print the top 10 words from the dictionary
print("Top 10 most occurring words from dictionary d:")
for word, count in top_10_words_from_d:
    print(f"{word}: {count}")
```
